```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load 
mishandled_baggage_to_load = "../Resources/Mishandled_Baggage.csv"
```

```python
# Read data
mishandled_baggage_df = pd.read_csv(mishandled_baggage_to_load)
```

```python
# Display the data table for preview
mishandled_baggage_df.head()
```

```python
# Convert enplaned_baggage to int
mishandled_baggage_df['ENPLANED_BAGGAGE'] = mishandled_baggage_df['ENPLANED_BAGGAGE'].astype('Int64')
display(mishandled_baggage_df.dtypes)
```

```python
# Get total row count per column
mishandled_baggage_df.count()
```

```python
# Clean the data
clean_mishandled_baggage_df = mishandled_baggage_df.drop(['PASSENGERS','CARRIER','CARRIER_NAME','PASSENGERS','MISHANDLED_WCHR_SCTR', 'ENPLANED_WCHR_SCTR','FORM_TYPE', 'MKT_CARRIER_FLAG'], axis=1, errors='ignore')
clean_mishandled_baggage_df.head()
```

```python
# Drop rows with NaN
final_mishandled_baggage_df=clean_mishandled_baggage_df.dropna() 
final_mishandled_baggage_df.head()
```

```python
# Get total row count per column
final_mishandled_baggage_df.count()
```

```python
# Get the total mishandled baggage per year
total_mishandled_per_year = final_mishandled_baggage_df.groupby(["YEAR"]).sum()["MISHANDLED_BAGGAGE"]
total_mishandled_per_year.head()
```

```python
# Get the total enplaned baggage per year
total_enplaned_per_year = final_mishandled_baggage_df.groupby(["YEAR"]).sum()["ENPLANED_BAGGAGE"]
total_enplaned_per_year.head()
```

```python
# Get the percentage of mishandled baggage per year
percentage_mishandled_per_year = (total_mishandled_per_year/ total_enplaned_per_year)*100
percentage_mishandled_per_year
```

```python
# Get the total mishandled baggage per month
total_mishandled_per_month = final_mishandled_baggage_df.groupby(["MONTH"]).sum()["MISHANDLED_BAGGAGE"]
total_mishandled_per_month
```

```python
# Get the total enplaned baggage per month
total_enplaned_per_month = final_mishandled_baggage_df.groupby(["MONTH"]).sum()["ENPLANED_BAGGAGE"]
total_enplaned_per_month
```

```python
# Get the percentage of mishandled baggage per month
percentage_mishandled_per_month = (total_mishandled_per_month/ total_enplaned_per_month)*100
percentage_mishandled_per_month
```

```python
percentage_mishandled_per_month.plot(figsize=(25,10),kind="line")
plt.ylabel('Percentage Mishandled', fontsize='30')
plt.xlabel('Month', fontsize='30')
plt.yticks(fontsize='25')
plt.xticks(fontsize='25')
plt.title('Percentage of Mishandled Baggage per Month', fontsize='40')
```

```python
# Get the total mishandled baggage per airline
total_mishandled_per_airline = final_mishandled_baggage_df.groupby(["UNIQUE_CARRIER_NAME"]).sum()["MISHANDLED_BAGGAGE"]
total_mishandled_per_airline
```

```python
# Get the total enplaned baggage per airline
total_enplaned_per_airline = final_mishandled_baggage_df.groupby(["UNIQUE_CARRIER_NAME"]).sum()["ENPLANED_BAGGAGE"]
total_enplaned_per_airline
```

```python
# Get the percentage of mishandled baggage per airline
percentage_mishandled_per_airline = (total_mishandled_per_airline/ total_enplaned_per_airline)*100
percentage_mishandled_per_airline
```

```python
percentage_mishandled_per_airline.plot(figsize=(25,10),kind="bar")
plt.ylabel('Percentage Mishandled', fontsize='30')
plt.xlabel('Airline', fontsize='30')
plt.yticks(fontsize='25')
plt.xticks(fontsize='25')
plt.title('Percentage of Mishandled Baggage per Airline', fontsize='40')
```
